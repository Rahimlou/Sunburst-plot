```python
import pandas as pd

df = pd.read_csv('FSXU0302R_Hierarchical_Kaiju_Taxonomy_Bacteria.csv')
```

```python
# Fill missing values with 'Not Specified'
df_filled = df.fillna('Not Specified')

# Print the DataFrame to check
print(df_filled)
```

```python
import plotly.express as px


# Create the Sunburst plot
fig = px.sunburst(df_filled,
                  path=['kingdom', 'phylum', 'order', 'family', 'genus'],
                  values='reads',
                  color = 'phylum')

# Show the plot
fig.show()
```

```python
fig.write_image("sunburst_plot_302R_Bacteria.pdf", format='pdf', width=800, height=600)
```
